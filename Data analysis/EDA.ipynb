categories MODEL, MAKE, MODEL YEAR

```python
#first we have to find the duplicates
data.duplicated(subset= ['Model', 'Make', 'Model Year'])
```

```python
#Now we have to invert the function, so we have the unique values and get the rows
~data.duplicated(subset= ['Model', 'Make', 'Model Year'])
unique_data = data.loc[~data.duplicated(subset= ['Model', 'Make', 'Model Year'])].reset_index(drop=True).copy()
unique_data
#We are not gonna work with unique_data 
```
